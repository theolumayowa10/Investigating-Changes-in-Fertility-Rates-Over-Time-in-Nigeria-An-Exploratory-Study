# Project Title
### Investigating Changes in Fertility Rates Over Time in Nigeria: An Exploratory Study

---

## Introduction

Nigeria’s demographic profile is rapidly evolving, with fertility behavior playing a central role in shaping population growth, healthcare needs, and socio‐economic planning. Although aggregate measures such as the Total Fertility Rate (TFR) offer a high‐level snapshot, age‐specific fertility rates (ASFR) reveal important nuances about when women are having children. By examining survey data spanning three decades (1990–2021), we can uncover patterns—such as teenage childbearing, peak fertility ages, and shifts in desired family size—that inform policy and program design in health, education, and economic development.

---

## 2. Problem Statement  
Despite periodic national surveys, there is limited consolidated analysis of how fertility patterns have changed over time across different age groups in Nigeria. Policymakers and public health practitioners lack a clear, data‐driven view of which age cohorts drive Nigeria’s high birth rates and how total and wanted fertility have evolved. This gap hinders targeted interventions—such as reproductive health education for adolescents or family‐planning services for peak‑fertility age groups—that could more effectively address maternal and child health outcomes.

---

## 3. Aims of the Project  
- **Characterize fertility trends** in Nigeria from 1990 to 2021 by:  
  - Age‐specific fertility rates (ASFR) for ages 15–49  
  - Total fertility rate (TFR) and wanted fertility rate  
  - General fertility rate (GFR) and crude birth rate (CBR)  
- **Identify age cohorts** with the highest fertility burden and assess changes over time  
- **Compare desired versus actual fertility**, highlighting any persistent gaps  
- **Provide evidence‑based recommendations** for targeted reproductive health programs  

---

## 4. Methodological Steps Taken  
1. **Data Acquisition**  
   - Imported the DHS‑derived CSV containing national fertility indicators for Nigeria (1990–2021)  

2. **Data Cleaning & Preparation**  
   - Removed the initial metadata row  
   - Converted key fields (`Value`, `SurveyYear`) to numeric types  
   - Dropped columns with more than 90% missing values  

3. **Descriptive Analysis**  
   - Computed overall summary statistics (mean, median, range)  
   - Tabulated average ASFR by age group and overall TFR metrics  

4. **Trend Visualization**  
   - Created line plots of ASFR for each age cohort over survey years  
   - Created line plots of TFR, wanted fertility rate, GFR, and CBR over time  

5. **Comparative Assessment**  
   - Overlaid actual vs. wanted fertility rates in the same chart  
   - Identified age cohorts with the largest gaps between desired and observed fertility  

6. **Interpretation & Recommendations**  
   - Synthesized key findings on declining or increasing trends by cohort  
   - Drafted targeted policy and programming suggestions based on age‑group insights

# Cleaning and Formatting 

### Removing first metadata row and reset index

```python
df = df.iloc[1:].reset_index(drop=True)
```

### Convert key columns to numeric

```python
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')
df['SurveyYear'] = pd.to_numeric(df['SurveyYear'], errors='coerce')
```

### Drop columns with >90% missing values

```python
df = df.loc[:, df.isnull().mean() < 0.9]
```

### Keep only essential columns

```python
df = df[['SurveyYear', 'Indicator', 'Value']].dropna()
```

# Prepare Data Subsets

```python
asfr_data = df[df['Indicator'].str.contains("Age specific fertility rate", na=False)]
tfr_data = df[df['Indicator'].str.contains("Total fertility rate", na=False)]
wanted_tfr = df[df["Indicator"] == "Total wanted fertility rate"]
gfr_data = df[df["Indicator"] == "General fertility rate"]
cbr_data = df[df["Indicator"] == "Crude birth rate"]
```

# Trend Variables

```python
tfr_trend = tfr_data.groupby("SurveyYear")["Value"].mean()
teen_fertility = asfr_data[asfr_data["Indicator"].str.contains("15-19")]
teen_trend = teen_fertility.groupby("SurveyYear")["Value"].mean()
gfr_trend = gfr_data.groupby("SurveyYear")["Value"].mean()
cbr_trend = cbr_data.groupby("SurveyYear")["Value"].mean()
```

# Actual vs wanted fertility

```python
merged_tfr = pd.merge(
    tfr_data.groupby("SurveyYear")["Value"].mean().reset_index(name="Actual_TFR"),
    wanted_tfr.groupby("SurveyYear")["Value"].mean().reset_index(name="Wanted_TFR"),
    on="SurveyYear"
)
```

# Age group breakdown

```python
avg_asfr = asfr_data.groupby("Indicator")["Value"].mean().sort_values(ascending=False)
asfr_pivot = asfr_data.pivot_table(index="SurveyYear", columns="Indicator", values="Value", aggfunc="mean")
asfr_change = asfr_pivot.iloc[-1] - asfr_pivot.iloc[0]
```

# Age specific group

```python
asfr_30_34 = asfr_data[asfr_data["Indicator"].str.contains("30-34")]
stable_age_group = asfr_30_34.groupby("SurveyYear")["Value"].mean()

asfr_45_49 = asfr_data[asfr_data["Indicator"].str.contains("45-49")]
older_fertility = asfr_45_49.groupby("SurveyYear")["Value"].mean()

asfr_15_24 = asfr_data[asfr_data["Indicator"].str.contains("15-19|20-24")]
early_childbearing = asfr_15_24.groupby("SurveyYear")["Value"].sum()
```

# Insights

## 1. Total Fertility Rate (TFR) Decline

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=tfr_trend.index, y=tfr_trend.values, marker='o')
plt.title("Total Fertility Rate (TFR) Over Time")
plt.xlabel("Survey Year")
plt.ylabel("Children per Woman")
plt.grid(True)
plt.show()
```

### - Total Fertility Rate (TFR) decreased from 5.85 in 1990 to 4.70 in 2021, indicating progress in family planning and awareness.


---

## 2. Teenage Fertility Rate (15–19)

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=teen_trend.index, y=teen_trend.values, marker='o', color='orange')
plt.title("Teenage Fertility Rate (15–19) Over Time")
plt.xlabel("Survey Year")
plt.ylabel("Births per 1,000 Women")
plt.grid(True)
plt.show()
```

### - Fertility among ages 15–19 declined from 146 to 103 births per 1,000, suggesting better reproductive education or access to contraception.


---

## 3. Actual vs. Wanted Fertility Comparison

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=merged_tfr, x='SurveyYear', y='Actual_TFR', label="Actual TFR", marker='o')
sns.lineplot(data=merged_tfr, x='SurveyYear', y='Wanted_TFR', label="Wanted TFR", marker='s')
plt.fill_between(merged_tfr['SurveyYear'], merged_tfr['Actual_TFR'], merged_tfr['Wanted_TFR'], alpha=0.3)
plt.title("Actual vs Wanted Fertility Over Time")
plt.xlabel("Survey Year")
plt.ylabel("Children per Woman")
plt.legend()
plt.grid(True)
plt.show()
```

### - The gap between actual and wanted fertility peaked at 0.45 children per woman in 2018, showing unmet contraceptive needs.

## 4. General Fertility Rate (GFR) Trend

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=gfr_trend.index, y=gfr_trend.values, marker='o', color='green')
plt.title("General Fertility Rate Over Time")
plt.xlabel("Survey Year")
plt.ylabel("Births per 1,000 Women (15–49)")
plt.grid(True)
plt.show()
```

### - GFR dropped from 207 (1990) to 164 (2021) births per 1,000 women aged 15–49.


---

## 5. Crude Birth Rate (CBR) Trend

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=cbr_trend.index, y=cbr_trend.values, marker='o', color='red')
plt.title("Crude Birth Rate Over Time")
plt.xlabel("Survey Year")
plt.ylabel("Births per 1,000 Population")
plt.grid(True)
plt.show()
```

### - CBR declined from 38.1 to 32.4 per 1,000 population, reflecting slowing population growth.


---    

## 6. Average Age-Specific Fertility Rates (ASFR)

```python
plt.figure(figsize=(10, 5))
avg_asfr.sort_values().plot(kind='barh', color='skyblue')
plt.title("Average Fertility Rate by Age Group (1990–2021)")
plt.xlabel("Average Births per 1,000 Women")
plt.ylabel("Age Group")
plt.grid(axis='x')
plt.show()
```

### - Women aged 25–29 had the highest average fertility rate (≈252.8), followed by 20–24 and 30–34.


---

## 7. Change in ASFR by Age Group

```python
plt.figure(figsize=(10, 5))
asfr_change.sort_values().plot(kind='barh', color='purple')
plt.title("Change in Fertility Rate by Age Group (1990 vs 2021)")
plt.xlabel("Change in Births per 1,000 Women")
plt.ylabel("Age Group")
plt.axvline(0, color='black', linestyle='--')
plt.grid(axis='x')
plt.show()
```

### All age groups saw a fertility drop; the sharpest in 20–24 (-46), 25–29 (-45), and 15–19 (-43) age groups.


---

## 8. Fertility Stability (Age 30–34)

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=stable_age_group.index, y=stable_age_group.values, marker='o', color='teal')
plt.title("Fertility Rate Trend (Age 30–34)")
plt.xlabel("Survey Year")
plt.ylabel("Births per 1,000 Women")
plt.grid(True)
plt.show()
```

### - Fertility in this age group stayed high but dropped from 244 (2003) to 193 (2021).

## 9. Older Age Fertility (Age 45–49)

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=older_fertility.index, y=older_fertility.values, marker='o', color='gray')
plt.title("Fertility Rate at Older Ages (45–49)")
plt.xlabel("Survey Year")
plt.ylabel("Births per 1,000 Women")
plt.grid(True)
plt.show()
```

### - Still present but low and fluctuating (e.g., 64 births per 1,000 in 1990 vs 43 in 2021).

## 10. Early Childbearing Risk (15–24 Combined)

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=early_childbearing.index, y=early_childbearing.values, marker='o', color='magenta')
plt.title("Combined Early Fertility (Age 15–24)")
plt.xlabel("Survey Year")
plt.ylabel("Total Births per 1,000 Women (15–24)")
plt.grid(True)
plt.show()
```

## 11. Fertility Between Older and Younger Female

### Filtering average fertility rates (ASFR) data

```python
asfr_young = asfr_data[asfr_data["Indicator"].str.contains("15-19|20-24")]
asfr_old = asfr_data[asfr_data["Indicator"].str.contains("40-44|45-49")]
```

### Calculating mean fertility rate for each group across all years

```python
young_mean = asfr_young["Value"].mean()
old_mean = asfr_old["Value"].mean()
```

### Preparing data for pie chart

```python
labels = ['Younger Females (15–24)', 'Older Females (40–49)']
sizes = [young_mean, old_mean]
colors = ['#66b3ff', '#ff9999']
```

### Plotting pie chart

```python
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Average Fertility Contribution: Younger vs Older Females")
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.show()
```

### - **Younger Females**: Ages **15–24**, representing teenagers and early adulthood.
### - **Older Females**: Ages **40–49**, representing the late reproductive age group.
### This comparison helps to highlight which demographic contributes more to overall fertility trends and may inform targeted reproductive health policies and programs.
